#### 8. Find the mean, median, and standard deviation of an array.

```python
import numpy as np
```

```python
np.random.seed(24)
```

```python
array = np.random.randint(1, 100, (2, 3))
print(array)
```

```python
mean = np.mean(array)
sum = np.sum(array)
sum_column = np.sum(array, axis=0)
sum_row = np.sum(array, axis=1)
median = np.median(array)
std = np.std(array)
```

```python
print("Sum of all elements : ", sum, end="\n")
print("Mean of all elements : ", mean, end="\n")
print("Median of all elements : ", median, end="\n")
print("Standard Deviation of all elements : ", std, end="\n")
print("Sum of all elements column wise : ", sum_column, end="\n")
print("Sum of all elements row wise : ", sum_row)
```
