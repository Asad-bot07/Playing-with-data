##### 10. Perform element-wise addition and multiplication on two arrays.

```python
import numpy as np
```

```python
np.random.seed(24)
```

```python
matrix = np.random.randint(1, 100, size=[2, 3])
vector = np.random.randint(1, 100, size=[1, 3])
```

```python
print(matrix)
print(vector)
```

```python
print(matrix + vector)
print(matrix * vector)
```
