### Find row-wise and column-wise sums of a 2D array.

```python
np.random.seed(24)
```

```python
array = np.random.randint(0, 100, size=[2, 3])
print(array)
```

```python
sum_column = np.sum(array, axis=0)
print(sum_column)
```

```python
sum_row = np.sum(array, axis=1)
print(sum_row)
```
