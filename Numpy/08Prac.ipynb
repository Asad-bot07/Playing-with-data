```python
import numpy as np
```

#### Creating an array

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr)
```

### Mathematical Operations

#### Sum of all the elements

```python
print("Sum of the array : ", np.sum(arr))
```

#### Mean of all the elements

```python
print("Mean : ", np.mean(arr))
```

#### Max of all the elements 

```python
print("Mean : ", np.mean(arr))
```

#### Min of all the elements

```python
print("Min : ", np.min(arr))
```

#### Standard deviation

```python
print("Standard deviation : ", np.std(arr))
```

#### Sums of rows

```python
print("Sum of rows : ", np.sum(arr, axis=1))
```

#### Sum of columns

```python
print("Sum of rows : ", np.sum(arr, axis=0))
```
