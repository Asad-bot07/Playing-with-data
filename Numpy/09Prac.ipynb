#### Boolean indexing
<p>boolean indexing means creating a boolean array using conditions and use that boolean array to pick only those element where the value is true</p>

#### Creating an 2d array

```python
arr = np.array([[1, 2, 3], [5, 7, 9]])
print(arr)
```

#### Filtering array with even elements

```python
evens = arr[arr % 2 == 0] 
print(evens)
```

<p>More filtering way</p>

```python
arr[arr > 3] = 0
print("Modified array : ", arr)
```
