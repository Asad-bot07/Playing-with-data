##### broadcasting : it allows numpy to perform arithmetic operations of diff shapes, smaller array matches the dimension of the larger array

<p> Creating an array </p>

```python
arr = np.array([[1, 2, 3]])
print(arr)
```

<p>Adding 10 to an array</p>

```python
print(arr + 10)
```

#### Creating an 2d array

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)
```

#### Creating an vector

```python
vector = np.array([9, 6 , 8])
print(vector)
```

#### Adding an 1d array and 2d array

```python
print(matrix + vector)
```
