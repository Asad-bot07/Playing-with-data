#### Creating an array

```python
arr = np.array([[1, 2, 3], [2, 3, 4]])
print(arr)
```

#### sum of the coulumn

```python
sum_column = np.sum(arr, axis=0)
print(sum_column)
```

#### sum of the rows

```python
sum_rows = np.sum(arr, axis=1)
print(sum_rows)
```
