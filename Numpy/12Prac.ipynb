```python
import numpy as np
```

#### Generating a random matrix

```python
matrix = np.random.randint(1, 51, size=[3,4])
print(matrix)
```

#### filtering data set

```python
matrix[matrix > 25] = 0
print("\n", matrix)
```

#### calculate summary stat

```python
print("Sum : ", np.sum(matrix))
print("Mean : ", np.mean(matrix))
print("Standard deviation : ", np.std(matrix))
```
