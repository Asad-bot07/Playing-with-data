```python
import numpy as np
```

### Mathematical operations on array

#### Creating basic arrays

```python
arr = np.array((1, 2, 3, 4, 5, 6, 10))

arr2 = np.array((2, 1, 5, 6, 7, 8, 9))
```

#### Addition

```python
print(arr + arr2)
```

#### Multiplication

```python
print(arr * arr2)
```

#### Subtraction

```python
print(arr - arr2)
```

#### Division

```python
print(arr / arr2)
```

### Creating another array with square values

```python
x = np.array((4, 9, 16, 25))
```

#### Square root

```python
print(np.sqrt(x))
```

#### sum of all elements

```python
print(np.sum(np.sqrt(x)))
```

#### mean of the array

```python
print(np.mean(x))
```

#### max of the array

```python
print(np.max(x))
```

#### min of the array

```python
print(np.min(x))
```
