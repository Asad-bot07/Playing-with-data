```python
import matplotlib.pyplot as plt
```

```python
years = [2021, 2022, 2023, 2024, 2025]
sales = [47, 82, 63, 109, 170]
```

#### Line Graph

```python
plt.plot(years, sales, label="Sales Trend", color="Red", marker="x")
plt.title("Sales as per 5 years")
plt.xlabel("Years")
plt.ylabel("Sales")
```

#### Bar Graph

```python
plt.bar(years, sales, color="red")
plt.title("Sales as per last 5 years")
plt.xlabel("Years")
plt.ylabel("Sales")
```

#### Histogram

```python
plt.hist(sales, color="red", bins=5, edgecolor="black")
```

#### Pie chart

```python
plt.pie(sales, labels=years, colors=["green", "lightgreen", "darkgreen", "lime", "forestgreen"], wedgeprops={'edgecolor': 'black'})
plt.title("Sales Distribution")
```

####  Scattered Graph

```python
plt.scatter(years, sales, color="red")
plt.title("Sales as per 5 years")
plt.xlabel("Years")
plt.ylabel("Sales")
```
