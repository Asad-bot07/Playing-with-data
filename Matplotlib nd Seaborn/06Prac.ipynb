```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

#### Seaborn : built on top of matplotlib for easier and more visually apealing plot, it provides a high level interfaces for statistical graphics eg. heatmap, pairplot 

##### Creating an random data

```python
data = np.random.randint(1, 10, size=[3, 5])
print(data)
```

##### Creating an heatmap

```python
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.title("Heatmap")
```

### Pairplot

```python
df = {
    "Name" : ["Asad", "Ayushi", "Rehman"],
    "Marks" : [38, 39, 14],
    "Criteria" : ["Pass", "Pass", "Fail"]
}
df = pd.DataFrame(df)
print(df)
```

```python
sns.pairplot(df)
```
