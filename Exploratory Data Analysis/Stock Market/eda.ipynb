# Stock Market Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Market.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] = df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].fillna(df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].interpolate())

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Date"], df["Close"], color='pink')
plt.title("Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Volume'])
plt.title("Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['Volume'], bins=6)
plt.title("Volume Distribution")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()
```

```python
df['high_low_diff'] = df['High'] - df['Low']
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['high_low_diff'])
```
