```python
import pandas as pd
```

### handling missing values

#####  Importing data set

```python
df = pd.read_csv('Data_set/MOCK_DATA.csv')
```

##### dropna : it will drop rows with missing values

```python
df = df.dropna()
```

###### this will drop rows with missing values

```python
df = df.dropna(axis=1)
```

##### this will fill the columns which have missing values with 0 or any numbers as we need the column and cannot just drop it

```python
df["column_name"] = df["column_name"].fillna(0)
```

#### forward fill (ffill) : fills missing values using the last valid value that appears before them

```python
f = df.fillna(method="ffill")
```

#### backward fill (bfil) : fills missing values using the next valid value that appears before them

```python
df = df.fillna(method="bfill")
```

#### interpolation way : fills missing vvalues in a straight line format

```python
df["column_name"] = df["column_name"].interpolate()
```
