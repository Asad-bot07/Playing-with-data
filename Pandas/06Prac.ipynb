```python
import pandas as pd
import numpy as np
```

#### creating a sample dataset

```python
data = {
    "name" : ["Asad", "Ayushi", "Taylor", np.nan, "Rehman"],
    "Age" : [20, 21, np.nan, 22, 34],
    "Score" : [np.nan, 95, np.nan, np.nan, 87]
}
```

#### Raw data to dataframe

```python
df = pd.DataFrame(data)
print("Original data : \n", df)
```

## cleaning data

```python
df = df.rename(columns={"name" : "first_name"})
print("Updated df : \n", df)
```

#### Interpolation filling

```python
df["Age"] = df["Age"].fillna(df["Age"].interpolate())
print("updated df :\n", df)
```

#### Filling missing values with the help of aggregation methods

```python

df["Score"] = df["Score"].fillna(df["Score"].mean())
```

```python
print("\nCleaned data : \n", df)
```
